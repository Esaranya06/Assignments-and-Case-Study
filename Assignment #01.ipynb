```python
#1.Read the dataset to python environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('iris.csv')
```

```python
# Display the columns in the dataset.
data.columns
```

```python
#Calculate the mean of each column of the dataset
np.mean(data[['SL', 'SW', 'PL', 'PW', 'Classification']])
```

```python
#Check for the null values present in the dataset.
data.isnull().sum()
```

```python
#Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.
#1. Histogram
plt.hist(data['PW'],color='red', rwidth=0.9, density= True, cumulative=True)
plt.title('Distribution of PW')
```

```python
#2. Swarm plot
sns.swarmplot(x='PW',y='Classification',data=data)
```

```python
#Boxplot
sns.boxplot(y='SL',x='Classification',data=data)
plt.ylabel('SL')
plt.xlabel('Classification')
```
